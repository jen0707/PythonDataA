```python
import matplotlib.pyplot as plt
import pandas as pd

netflix = pd.read_csv('netflix_cleaned.csv')

#시리즈 타입
#groupby()로 'type'열끼리 묶기, 'duration' 평균 구하기
netflix.groupby('type')[['duration']].mean()
```

```python
#연도별 작품의 수: 개수 세기 (count)
netflix.groupby('release_year')[['show_id']].count()
```

```python
#데이터프레임 타입
#groupby로 'type', 'country' 열로 묶기, 'release_year', 'duration' 열 평균 구하기
#plot으로 그래프 그리기
netflix.groupby(['type','country'])[['release_year', 'duration']].mean().reset_index().sort_values(by='duration', ascending=False).reset_index(drop=True).plot(kind='bar', x='country', y='duration', figsize=(15, 5))


plt.xticks(rotation=30)
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Square')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100) #numpy 배열 생성: linspace() / 1부터 10까지 균일한 간격으로 100개 생성
y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = 2x')
plt.plot(x, y3, label='y = 0.5x')
plt.plot(x, y4, label='y = 3x')

plt.legend(loc='upper right', ncol=2) #위치설정(loc, 들어갈 수 있는 문자가 정해져있으므로 참고할 것), n줄씩(ncol)

plt.title('Simple Linear Plot Example')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [3, 6, 9, 12])

plt.xlim([0, 5]) #x축 범위 지정
plt.ylim([0, 15]) #y축 범위 지정

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [3, 6, 9, 12])

plt.axis([0, 5, 0, 15]) #axis([xmin, xmax, ymin, ymax]): 범위 지정을 한번에
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.exp(x) #지수함수

plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Linear Scale')

plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.yscale('log') #y축만 log scale
plt.title('Logarithmic Scale (y-axis)')

plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.xscale('log')  # x축만 log scale
plt.title('Log-Linear Scale (x-axis)')

plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.yscale('log')     # y축 log scale
plt.xscale('linear')  # x축 선형 (default; 생략 가능)
plt.title('Linear-Logarithmic Scale')

plt.tight_layout() #레이아웃 자동 조정
plt.show()
```

```python
#선 종류 설정_label
plt.plot([1, 2, 3], [4, 4, 4], '-', label='solid')
plt.plot([1, 2, 3], [3, 3, 3], '--', label='dashed')

#선 종류 설정_linestyle
plt.plot([1, 2, 3], [2, 2, 2], linestyle='dotted', label='dotted')
plt.plot([1, 2, 3], [1, 1, 1], linestyle='dashdot', label='dash-dot')


plt.axis([0.8, 3.2, 0.5, 5.0])
plt.legend(loc='upper center', ncol=4)
plt.show()
```

```python
plt.plot([4, 5, 6], "b") #blue
plt.plot([3, 4, 5], "ro") #red circle
plt.plot([2, 3, 4], marker = "s") #square
plt.plot([1, 2, 3], marker = "D") #diamond
plt.plot([0, 1, 2], marker = '$M$') #$문자$: 사용자 지정
plt.show()
```

```python
#색상 설정
#이름
plt.plot([1, 2, 3, 4], [2.0, 3.3, 6.3, 10.5], color = 'violet')

#약자
plt.plot([1, 2, 3, 4], [2.0, 3.1, 5.3, 8.5], color = 'g')

#RGB
plt.plot([1, 2, 3, 4], [2.0, 2.8, 4.3, 6.5], color=(0.1, 0.2, 0.3))

#16진수
plt.plot([1, 2, 3, 4], [2.0, 2.5, 3.3, 4.5], color='#FF0000')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]

#첫 번째 서브 그래프
plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Graph Sub Title 1', loc = 'left')

#두 번째 서브 그래프
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Graph Sub Title 2', loc = 'left')

#슈퍼 타이틀 설정
plt.suptitle('Graph Super Title')

#간격 조절
plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3]

years = ['2030', '2040', '2050']
values = [300, 100, 700]

#막대그래프
plt.bar(x, values, color = ['r', 'g', 'b'])

#x축 눈금에 '2030', '2040', '2050'를 표시
plt.xticks(x, years)
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 5, 10, 5])

#그리드 활성화, 설정
#linestyle: 선 스타일, linewidth: 선 굵기, color: 선 색상, alpha: 투명도
#axis: x축을 지정하면 x축에, y축을 지정하면 y축에 해당되는 그리드(default; x축 y축 같이)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 20]

plt.bar(categories, values, color='skyblue')

#enumerate(), for문을 통해 각 막대 위에 값 텍스트로 표시
#x좌표 : i / y좌표 : value + 1
#str : 문자열 변환 / ha, va : 정렬

#plt.test(x좌표, y좌표, 내용) : 텍스트 표시
for i, value in enumerate(values):
  plt.text(i, value + 1, str(value), ha = 'center', va = 'bottom', color = 'black', fontsize = 10, fontweight = 'bold' )

plt.title('Bar Chart with Text Annotations')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 20]

plt.bar(categories, values, color='skyblue')

#각 막대의 내부 x좌표 출력
for i, value in enumerate(values):
    plt.text(i, value + 1, f"x={i}", ha='center', color='red')

plt.grid(True) #grid
plt.show()
```

```python
import pandas as pd

titanic = pd.read_csv('titanic_selena.csv')
```

```python
#첫 5개 행 출력
titanic.head()
```

```python
#열 요약 정보 확인
titanic.info()
```

```python
#객실 등급에 따른 생존자와 사망자의 평균
pclass_survived_mean = titanic.groupby('Pclass')['Survived'].mean().reset_index()
pclass_survived_mean
```

```python
plt.plot(pclass_survived_mean['Pclass'], pclass_survived_mean['Survived'], marker='o', linestyle='-', color='violet')
plt.title('Survival Rate Variation Across Passenger Classes')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.xticks([1, 2, 3])
plt.grid(True)
plt.show()
```

좋은 자리일 수록 생존률이 높다는 것을 알 수 있다.

```python
titanic['Pclass'].value_counts()
```

```python
#승선 항구에 따른 생존자의 수
survived_counts = titanic[titanic['Survived'] == 1]['Embarked'].value_counts()
survived_counts
```

```python
#막대그래프 그리기
plt.bar(survived_counts.index, survived_counts, color = ['mediumorchid', 'darkviolet', 'indigo']) #X, Y, 색 지정
plt.title('Survived Counts by Embarked Port on Titanic')
plt.xlabel('Embarked Port')
plt.ylabel('Count')
plt.xticks(survived_counts.index, ['Southampton', 'Cherbourg', 'Queenstown'])
plt.legend(['Survived'], loc='upper right') #찾은 값 표시, loc 위치 지정
plt.grid(axis='y', linestyle='--', alpha=0.7) #grid 표시

#생존자 수 표시
for i, value in enumerate(survived_counts):
  plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.show()
```

```python
#성별에 따른 생존자 수 계산
survived_counts = titanic[titanic['Survived'] == 1]['Sex'].value_counts()
survived_counts
```

```python
#그래프를 변수에 지정
bars = plt.barh(survived_counts.index, survived_counts, color = ['darkturquoise', 'salmon'])

plt.title('Survived Counts by Gender on Titanic')
plt.xlabel('Count')
plt.ylabel('Gender')

plt.legend(bars, ['Survived - Feamle', 'Survived - Male'], loc='upper right') #범례

#수직선 추가: plt.axvline
plt.axvline(x=survived_counts['male'], color='gray', linestyle='--', linewidth=1)

#수평선: plt.axhline
plt.axhline(y=0.5, color='gray', linewidth=1)

for i, value in enumerate(survived_counts):
  plt.text(value + 1, i, str(value), ha='left', va='center')

plt.show()
```

```python
titanic = pd.read_csv('titanic_selena.csv')
print(titanic.info(), '\n')
```

```python
#결측치 처리
titanic = titanic.dropna(subset=['Age', 'Fare', 'Survived'])
print(titanic.info())
```

```python
plt.figure(figsize=(12, 8)) #그래프 크기 지정 생성: 12인치 x 8인치
scatter = plt.scatter(x='Age', y='Fare', data=titanic, c=titanic['Survived'],  # c: 생존여부에 따라 색 다르게 지정
                      cmap='Set2', alpha=0.7) # Cmap: 색상 세트 지정

plt.title('Age and Fare Relationship with Survival on the Titanic')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.legend(handles=scatter.legend_elements()[0], title='Survived',
           labels=['Not Survived', 'Survived'], loc='upper right')
plt.show()
```

```python
#seaborn library를 쓰면 더 간편하게 나타낼 수 있음
import seaborn as sns

sns.scatterplot(data= titanic, x='Age',
                 y='Fare',
                 hue='Survived', palette='Set2', alpha=0.5)

plt.title('Age and Fare Relationship with Survival on the Titanic')
plt.xlabel('Age')
plt.ylabel('Fare')
```
