```python
#for문으로 결측치 확인
for i in netflix.columns :
  missingValueRate = netflix[i].isna().sum() / len(netflix) * 100
  if missingValueRate > 0 :
    print("{} null rate: {}%".format(i, round(missingValueRate, 2))) #소수 둘째자리 반올림
```

```python
netflix["country"] = netflix["country"].fillna("No Data") #결측치를 No Data로 대체
netflix
```

```python
netflix["director"] = netflix["director"].replace(np.nan, "No Data") #np.nan(결측치)를 No Data로 대체
netflix
```

```python
netflix.dropna(axis=1, inplace=True) #axis(줄) = 1(세로: numpy), inplace=True 해줘야 적용됨
netflix
```

```python
netflix = pd.read_csv("netflix_selena.csv")

netflix.dropna(subset=["listed_in"], inplace=True) #listed_in 열에서 결측치가 존재하는 항만 제거
netflix
```

```python
#실습용 전처리
netflix = pd.read_csv("netflix_selena.csv")

netflix["country"] = netflix["country"].fillna("No Data")
netflix["director"] = netflix["director"].replace(np.nan, "No Data")
netflix.dropna(axis=1, inplace=True)
netflix
```

```python
netflix.to_csv("netflix_cleaned.csv", index=True)
```

```python
netflix = pd.read_csv("netflix_cleaned.csv")

print("mean: ", netflix["duration"].mean()) #평균값
print("median: ", netflix["duration"].median()) #중앙값
print("sum: ", netflix["duration"].sum()) #합
print("min: ", netflix["duration"].min()) #최솟값
print("max: ", netflix["duration"].max()) #최댓값
```

```python
print("var : ", netflix["duration"].var()) #분산
print("std : ", netflix["duration"].std()) #표준편차
```

```python
print("count :", netflix["type"].count()) #값의 개수
print("value_counts :", netflix["type"].value_counts()) #빈도수
```

```python
print(netflix[["release_year", "duration"]].describe()) #통계 지표 요약
```

```python
print(netflix.agg({"release_year" : ["min", "max", "median", "std"],
                   "duration" : ["min", "max", "mean", "median"]}))
```

```python
#원하는 통계량
print(netflix.agg({"release_year" : ["min", "max", "median", "std", "mean"],
                   "duration" : ["min", "max", "mean", "median"]}))
```
